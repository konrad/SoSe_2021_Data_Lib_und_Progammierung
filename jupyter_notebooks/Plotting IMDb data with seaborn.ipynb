```python
# mit der Methode apply kann eine Funktion auf alle Werte eine Series angewandt werden;
# es wird eine neue Series zurückgegeben, die die Rückgabewerte der Funktion enthält
def multiply_by_1000(number):
    return "blub-" + str(number * 1000)

imdb.averageRating.apply(multiply_by_1000)
```

```python
# Lambda Funktionen (Unbennante Funktionen)

imdb.averageRating.apply(lambda number: "blub-" + str(number * 1000))
```

```python
# Wir generieren eine neue Spalte (Series), in der wir unsere Filme in zwei Klasse einteilen. 
# Solche die 20 (der Median) oder mehr Stimmen bekommen haben und solche
# die weniger Stimmen bekommen haben.
```

```python
imdb_subsample["voting_class"] = imdb_subsample.numVotes.apply(
    lambda number: "more_or_equal_20" if number >= 20 else "less_than_20"
)
```

```python
imdb_subsample
```

```python
sns.stripplot(data=imdb_subsample, y="averageRating", x="voting_class", alpha=0.1)
```

```python
sns.boxplot(data=imdb_subsample, y="averageRating", x="voting_class")
```
